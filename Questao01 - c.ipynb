```python
#Questão01 - letra c - falsa posição.

listaA = []
listaB = []
listaC = []
listaba = []
contador = 0

precisao = float(input("Precisão: "))

a = float(input("Valor de a: "))
b = float(input("Valor de b: "))

def ProfCritica(y):
    
    A = (3*y) + ((y**2)/2)
    B = 3+y
    g = 9.81
    Q = 20
    
    f = 1 - (((Q**2) / (g*(A**3))) * B )
    
    return f

while(abs(b-a) > precisao):
    
    fa = ProfCritica(a)
    fb = ProfCritica(b)
    
    c = a - (((b-a)/(fb-fa))*fa)
    
    fc = ProfCritica(c)
    
    if (fc*fa) < 0:
        b=c
        a=a
    else:
        a=c
        b=b
    
    contador = contador + 1

    print("iteração: {} / valor atual de c: {}: ".format(contador, c))

print("\nValor final de c: ", c)
    
```
