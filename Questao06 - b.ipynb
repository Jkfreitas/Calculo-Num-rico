```python
#      c0   c1   c2   c3    b
#
# l0   22  -7    0   -21 | -60
# l1   7    65  -42  -30 |  0
# l2   0    18  -13  -5  |  0
# l3   63   90   35  -62 | 25200 

#      c0   c1   c2   c3
#
# l0   63  90   35  -62
# l1   7   65  -42  -30     
# l2   0   18  -13  -5     
# l3   22  -7   0   21

#---------------------------------------------------------------------------------------------

MatrizPivo = []
MatrizPermutacao = []
MatrizA = [[22, -7, 0, 21], [7, 65, -42, -30], [0, 18, -13, -5], [63, 90, 35, -62]]
CopiaMatrizA = [[22, -7, 0, 21], [7, 65, -42, -30], [0, 18, -13, -5], [63, 90, 35, -62]] #cópia usada para checar quais linhas foram trocada (não da pra usar a original porque apos trocar as linhasa original é perdida)
ListaPivo = []

#---------------------------------------------------------------------------------------------

def search (lista, valor):
    for m in lista:
        if valor in m:
            return lista.index(m)


def DefineMaiorAbs(x, y, z): #Função que retorna a posição do maior valor absoluto da MatrizPivo
    if(abs(x) > abs(y)):
        return z.index(x)
    else:
        return z.index(y)
    
def DefineMaiorValor(valor1, valor2): #Função que retorna o maior valor absoluto da MatrizPivo
    if(abs(valor1) > abs(valor2)):
        return valor1
    else:
        return valor2 
    
#-----------------------------------------------------------------------------------------------
 
for i in range(0, 4):            #Leitura das linhas
    for j in range(i, 4):        #Leitura das colunas
        aux = MatrizA[j][i]      #Armazena todos os valores da coluna em aux
        MatrizPivo.append(aux)   #Preenche MatrizPivo com os valores das colunas
        
    valorPivo = DefineMaiorValor(max(MatrizPivo), min(MatrizPivo)) 
    Permuta = search(CopiaMatrizA, valorPivo)
    ListaPivo.append(valorPivo)
    print("Valor pivo: ", valorPivo)
    print("Permuta: ", Permuta)
    print("Matriz Pivo: ", MatrizPivo)
    print("Posição do maior Abs: ", DefineMaiorAbs(max(MatrizPivo), min(MatrizPivo), MatrizPivo))
    print("----------------------------------------------------------------------------------------")
    
    pivo = DefineMaiorAbs(max(MatrizPivo), min(MatrizPivo), MatrizPivo) #Armazena o resultado de DefineMaiorAbs em pivo
    
    MatrizPermutacao.append(Permuta) #Matriz Permutação
    
    MatrizA[i], MatrizA[pivo+i] = MatrizA[pivo+i], MatrizA[i] #Troca a linha atual pelo pivo
        
    del MatrizPivo[0:]
    
    print("Matriz A: ", MatrizA)
    
print("Matriz de permutação: ", MatrizPermutacao)
print("----------------------------------------------------------------------------------------")

#-----------------------------------------------------------------------------------------------------

MatrizU = [[63, 90, 35, -62], [7, 65, -42, -30], [0, 18, -13, -5], [22, -7, 0, 21]]
itemY = 0

for itemX in range(len(MatrizU)):
    mult= MatrizU[itemX][itemY]/MatrizU[itemX][itemX]
    
    for itemY in range(len(MatrizU)):
        if (itemX==0):
            break
        MatrizU[itemX][itemY] = MatrizU[itemY][itemX] - (MatrizU[itemX][itemY]*mult)

print("\n\n")
print("Matriz U:", MatrizU)



            




```
